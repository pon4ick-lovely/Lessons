### Основные команды (подробнее)
- **`ls`**: Просмотр содержимого директории.
  - `ls -l`: Подробный список файлов с правами доступа, владельцем, размером и датой изменения.
  - `ls -a`: Показывает скрытые файлы.
  - `ls -lh`: Показывает размеры файлов в удобочитаемом формате (например, KB, MB).
- **`pwd`**: Показ текущей рабочей директории.
- **`cd`**: Переход в директорию.
  - `cd /path/to/directory`: Переход в указанную директорию.
  - `cd ..`: Переход на уровень выше.
  - `cd ~`: Переход в домашнюю директорию пользователя.
  - `cd -`: Возвращение в предыдущую директорию.
- **`mkdir`**: Создание новой директории.
  - `mkdir new_directory`: Создает директорию `new_directory`.
  - `mkdir -p /path/to/directory`: Создает директорию, включая промежуточные, если они не существуют.
- **`rm`**: Удаление файла или директории.
  - `rm filename`: Удаляет файл `filename`.
  - `rm -r directory_name`: Удаляет директорию и её содержимое.
  - `rm -f filename`: Принудительное удаление файла (без запроса подтверждения).
- **`cp`**: Копирование файла или директории.
  - `cp source_file destination_file`: Копирует файл.
  - `cp -r source_directory destination_directory`: Копирует директорию и её содержимое.
  - `cp -p source_file destination_file`: Копирует файл, сохраняя права доступа и временные метки.
- **`mv`**: Перемещение или переименование файла или директории.
  - `mv old_name new_name`: Переименование файла или директории.
  - `mv file_name /path/to/directory`: Перемещение файла в другую директорию.
- **`touch`**: Создание пустого файла.
  - `touch filename`: Создает пустой файл `filename`.
  - `touch -c filename`: Обновляет временные метки файла, если он существует.
- **`cat`**: Просмотр содержимого текстового файла.
  - `cat filename`: Печатает содержимое файла `filename`.
  - `cat file1 file2 > combined_file`: Объединяет два файла в один.
- **`echo`**: Вывод текста в терминал.
  - `echo "Hello, World!"`: Печатает "Hello, World!" в терминале.
  - `echo "text" > file.txt`: Записывает текст в файл (перезаписывает файл).
  - `echo "text" >> file.txt`: Дописывает текст в конец файла.
- **`ln`**: Создание ссылок на файлы.
  - `ln source_file link_name`: Создает жесткую ссылку.
  - `ln -s source_file link_name`: Создает символическую ссылку (ярлык).

### Команды для работы с файлами:
- **`head`**: Просмотр начала файла.
  - `head filename`: Выводит первые 10 строк файла.
  - `head -n 20 filename`: Выводит первые 20 строк файла.
- **`tail`**: Просмотр конца файла.
  - `tail filename`: Выводит последние 10 строк файла.
  - `tail -n 20 filename`: Выводит последние 20 строк файла.
  - `tail -f filename`: Непрерывно выводит новые строки файла (полезно для логов).
- **`less`**: Просмотр содержимого файла постранично.
  - `less filename`: Открывает файл для постраничного просмотра.
- **`more`**: Просмотр содержимого файла постранично (аналог `less`).
  - `more filename`: Открывает файл для постраничного просмотра.
- **`grep`**: Поиск текста в файле или директории.
  - `grep "pattern" filename`: Ищет строку, содержащую "pattern", в файле.
  - `grep -r "pattern" directory`: Рекурсивно ищет в директории.
  - `grep -i "pattern" filename`: Ищет без учета регистра.
  - `grep -v "pattern" filename`: Выводит строки, не содержащие "pattern".
- **`find`**: Поиск файлов и директорий.
  - `find /path/to/search -name "filename"`: Ищет файл с именем `filename`.
  - `find /path/to/search -type d -name "directory_name"`: Ищет директорию с именем `directory_name`.
  - `find /path/to/search -mtime -1`: Ищет файлы, измененные за последние 24 часа.
- **`wc`**: Подсчет слов, строк и символов в файле.
  - `wc filename`: Выводит количество строк, слов и байтов в файле.
  - `wc -l filename`: Подсчитывает количество строк.
  - `wc -w filename`: Подсчитывает количество слов.

### Команды для управления процессами:
- **`ps`**: Просмотр списка запущенных процессов.
  - `ps aux`: Полный список процессов.
  - `ps -ef`: Альтернативный формат отображения списка процессов.
- **`kill`**: Завершение процесса.
  - `kill PID`: Завершает процесс с указанным PID.
  - `kill -9 PID`: Принудительное завершение процесса.
- **`top`**: Мониторинг активных процессов в реальном времени.
  - `top`: Показывает активные процессы в реальном времени.
  - Внутри `top`: Нажмите `q`, чтобы выйти, `h` для справки, `M` для сортировки по использованию памяти.
- **`df`**: Проверка доступного дискового пространства.
  - `df -h`: Показ свободного места на дисках в удобном для чтения формате.
- **`du`**: Показ размера директорий и файлов.
  - `du -sh directory_name`: Показ размера директории в удобочитаемом формате.
  - `du -sh *`: Показ размера всех файлов и папок в текущей директории.

### Дополнительные команды:
- **`history`**: Просмотр истории выполненных команд.
  - `history`: Печатает список всех ранее введенных команд.
  - `!number`: Повторить команду из истории с указанным номером.
- **`clear`**: Очистка терминала.
  - `clear`: Очищает экран терминала.
- **`man`**: Просмотр документации по командам.
  - `man command`: Показывает справку по команде `command`.
- **`info`**: Альтернативная команда для просмотра документации.
  - `info command`: Показывает более детализированную справку по команде `command`.
- **`sudo`**: Выполнение команд с правами суперпользователя.
  - `sudo command`: Выполняет команду с привилегиями администратора.
- **`uptime`**: Показ времени работы системы с момента последнего запуска.
  - `uptime`: Показывает, как долго система работает, сколько пользователей и средняя загрузка.
- **`whoami`**: Показ имени текущего пользователя.
  - `whoami`: Выводит имя пользователя, под которым выполнен вход.
- **`hostname`**: Показ имени хоста системы.
  - `hostname`: Выводит имя хоста (компьютера).
- **`date`**: Показ текущей даты и времени.
  - `date`: Показывает текущую дату и время.
  - `date "+%Y-%m-%d %H:%M:%S"`: Форматированный вывод даты и времени.
- **`cal`**: Показ календаря.
  - `cal`: Показывает календарь на текущий месяц

## Как можно работать с окружениями

### Ресурсы

- [Виртуальное окружение - step на курсе Телеграм-боты на Python и AIOgram](https://stepik.org/lesson/759386/step/1?unit=761402)

Для создания окружений в Python обычно используются виртуальные окружения. Это позволяет изолировать зависимости разных проектов и избежать конфликтов между ними. Вот основные способы создания виртуальных окружений в Python:

### 1. Использование `venv`

`venv` — это встроенный модуль в Python (начиная с версии 3.3), который позволяет создавать виртуальные окружения.

#### Создание виртуального окружения с использованием `venv`:

1. Откройте терминал или командную строку.
2. Перейдите в папку вашего проекта.
3. Создайте виртуальное окружение командой:

   ```bash
   python -m venv myenv
   ```

   Здесь `myenv` — это имя вашего виртуального окружения. Вы можете использовать любое другое имя.

4. Активируйте виртуальное окружение:

   - На Windows:

     ```bash
     myenv\Scripts\activate
     ```

   - На macOS/Linux:

     ```bash
     source myenv/bin/activate
     ```

5. Теперь вы можете устанавливать пакеты, и они будут установлены только в этом окружении:

   ```bash
   pip install <package_name>
   ```

6. Чтобы деактивировать окружение, используйте команду:

   ```bash
   deactivate
   ```

### 2. Использование `virtualenv`

`virtualenv` — это внешняя библиотека, которая также позволяет создавать виртуальные окружения. Этот метод может быть полезен для совместимости со старыми версиями Python.

#### Установка и использование `virtualenv`:

1. Установите `virtualenv`, если он не установлен:

   ```bash
   pip install virtualenv
   ```

2. Перейдите в папку проекта и создайте виртуальное окружение:

   ```bash
   virtualenv myenv
   ```

3. Активируйте окружение так же, как и с `venv` (см. шаг 4 выше).

### 3. Использование `conda` (если у вас установлены Anaconda/Miniconda)

`conda` — это пакетный менеджер и система управления окружениями, которая устанавливается вместе с Anaconda или Miniconda.

#### Создание окружения с `conda`:

1. Откройте терминал или Anaconda Prompt.
2. Создайте окружение с конкретной версией Python:

   ```bash
   conda create --name myenv python=3.8
   ```

3. Активируйте окружение:

   ```bash
   conda activate myenv
   ```

4. Для установки пакетов используйте `conda install` или `pip`.

5. Чтобы деактивировать окружение, выполните:

   ```bash
   conda deactivate
   ```

### 4. Использование `pipenv`

`pipenv` — это инструмент, который объединяет управление зависимостями и создание виртуальных окружений.

#### Установка и использование `pipenv`:

1. Установите `pipenv`, если он не установлен:

   ```bash
   pip install pipenv
   ```

2. Перейдите в папку проекта и создайте окружение:

   ```bash
   pipenv install
   ```

   Это создаст виртуальное окружение и `Pipfile` для управления зависимостями.

3. Чтобы активировать виртуальное окружение:

   ```bash
   pipenv shell
   ```

4. Для установки новых зависимостей используйте:

   ```bash
   pipenv install <package_name>
   ```

Выбор метода зависит от ваших предпочтений и конкретных требований проекта. `venv` и `conda` — это наиболее популярные и часто используемые инструменты.

## Аннотация типов и документация 

````python
from typing import List


def greet(name: str) -> str:
    """
    Приветствует пользователя.

    Args:
        name (str): Имя пользователя

    Returns:
        str: Приветствие пользователя
    """
    return f"Привет, {name}!"

def sum_numbers(a: int, b: int) -> int:
    """
    Суммирует два числа.

    Args:
        a (int): Первое число
        b (int): Второе число

    Returns:
        int: Сумма двух чисел
    """
    return a + b

def calculate_average(numbers: List[int]) -> float:
    """
    Вычисляет среднее значение списка чисел.

    Args:
        numbers (List[int]): Список чисел

    Returns:
        float: Среднее значение списка
    """
    return sum(numbers) / len(numbers)

def process_string(input_str: str) -> str:
    """
    Преобразует строку в верхний регистр и добавляет "!" в конец.

    Args:
        input_str (str): Входная строка

    Returns:
        str: Модифицированная строка
    """
    return input_str.upper() + "!"

# Использование функций
print(greet("Мир"))  # Привет, Мир!
print(sum_numbers(5, 3))  # 8
numbers = [1, 2, 3, 4, 5]
average = calculate_average(numbers)
print(f"Среднее значение: {average}")  # Среднее значение: 3.0
result = process_string("Привет")
print(result)  # ПРИВЕТ!
````

## Интроспекция

Интроспекция в Python - это процесс изучения внутренней структуры объекта или класса. Она позволяет вам получить информацию о типе объекта, его атрибутах и методах.

Чтобы использовать интроспекцию в Python, вам нужно знать следующие вещи:

1. Типы данных: Python поддерживает различные типы данных, такие как строки, числа, списки, словари и т.д. Вы можете использовать функцию type() для определения типа данных объекта.

2. Атрибуты: Атрибуты - это свойства объекта, которые могут быть как встроенными, так и пользовательскими. Вы можете получить список всех атрибутов объекта с помощью функции dir().

3. Методы: Методы - это функции, которые принадлежат объекту и могут быть вызваны через точку (.). Вы можете получить список всех методов объекта с помощью функции dir().

4. Наследование: Классы в Python могут наследоваться друг от друга. Вы можете узнать, от какого класса наследуется объект, используя функцию isinstance().

5. Переопределение методов: Вы можете переопределить методы класса, чтобы изменить их поведение. Это может быть полезно для расширения функциональности класса без изменения исходного кода.

6. Метаклассы: Метаклассы в Python используются для управления созданием классов и объектов. Вы можете использовать метаклассы для добавления новых атрибутов или методов к классу.

7. Модули: Модули в Python - это файлы, содержащие код Python. Вы можете импортировать модули и использовать их функции и классы в своем коде.

8. Импортирование: Вы можете импортировать модули и классы в свой код с помощью ключевого слова import.

9. Расширения: Вы можете расширить функциональность Python с помощью внешних библиотек и модулей.

10. Безопасность: При использовании интроспекции важно помнить о безопасности, так как злоумышленник может использовать интроспекцию для получения информации о вашем приложении.


### Пример для строки

````python
a: int = 'string a'

print(type(a))  # <class 'str'>
print(isinstance(a, str))  # True
print(getattr(a, '__class__', None))  # <class 'str'>
# print(vars(a))  # {'__class__': <class 'str'>, '__del__': <method 'del' of '_str' objects>, ...}
print(repr(a))  # 'string a'
print(str(a))  # string a
print(dir(a))  # ['__add__', '__class__', '__contains__', ...]
print(hasattr(a, '__class__'))  # True
````

### Пример для функции

````python
import inspect # <------ библиотека inspect


def greet(name: str) -> str:
    """
    Приветствует пользователя.

    Args:
        name (str): Имя пользователя

    Returns:
        str: Приветствие пользователя
    """
    return f"Привет, {name}!"


def get_function_info(func, *args):
    annotations = func.__annotations__
    docstring = inspect.getdoc(func) # func.__doc__
    
    print(f"Функция: {func.__name__}")
    print(f"Документация:")
    print(docstring)
    print("\nАргументы:")
    for arg_name, arg_type in annotations.items():
        print(f"- {arg_name}: {arg_type}")
    
    return_value = func(*args)
    print(f"\nВозвращаемое значение: {type(return_value).__name__}")

# Пример использования
get_function_info(greet, 'example')
````
